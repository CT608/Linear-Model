```r
library(tidyverse)
library(skimr)
library(ggplot2)
```

```r
#data processing----------------------------------------------------------------
#import data
data = read.csv("https://raw.githubusercontent.com/CT608/Linear-Model/main/Air%20pollution%20and%20health%20project%202.csv")

#calculate rate
data$rate_reli = data$Yreli / data$Ereli
data$rate_prev = data$Yprev / data$Eprev

#convert month type as date type
data$date = paste(data$month, "01", sep="-")
data$date = format(as.Date(data$date, "%b-%y-%d"),format="%Y-%m")

#factor discrete data
data$board = as.factor(data$board)
data$dispensing = as.factor(data$dispensing)
data$urban = as.factor(data$urban)

#multiply easting and northing by 1000 coordinates
data$easting = data$easting*1000
data$northing = data$northing*1000
```

```r
data
```

```r
#time series rend line
data_ts = data %>%
  group_by(date) %>%
  summarize(Yreli = sum(Yreli),
            Yprev = sum(Yprev),
            Ereli = sum(Ereli),
            Eprev = sum(Eprev),) %>%
  mutate(rate_reli = Yreli / Ereli,
         rate_prev = Yprev / Ereli)

ggplot(data=data_ts) +
  geom_line(aes(x=date, y=rate_reli, group=1)) +
  geom_point(aes(x=date, y=rate_reli))

ggplot(data=data_ts) +
  geom_line(aes(x=date, y=rate_prev, group=1)) +
  geom_point(aes(x=date, y=rate_prev))


ggplot(data, aes(x = date, y = rate_reli, group = factor(code))) +
  geom_line() +
  geom_point(size = 2) +
  theme_bw() + theme(legend.position = "none")
```
